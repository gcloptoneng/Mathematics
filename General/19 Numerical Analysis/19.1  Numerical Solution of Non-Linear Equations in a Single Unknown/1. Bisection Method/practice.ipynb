Make plot for $f(x) = x^3 + 4x^2 - 10$, and use it to find the range in which the solution lies. Then use the bisection method to estimate the solution to a tolerance of $1 \times 10^{-5}$. Find the roots of the polynomial using np.roots and compare the result to that of the bisection method.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.linspace(-2, 2, 400)
f = x**3 + 4*x**2 - 10

plt.figure(figsize=(8, 6))
plt.plot(x, f, color = "cyan", linewidth = 0.5)
plt.axvline(0, color = "yellow", linewidth = 0.5)
plt.axhline(0, color = "yellow", linewidth = 0.5)
plt.grid(color = "gray", linestyle = "--", linewidth = 0.5)
plt.show()
```

```python
def f(x):
    return x**3 + 4*x**2 - 10

def bisection_method(a, b, tol, max_iter):
    if f(a)*f(b) >= 0:
        print("The bisection method failed.")
        return None

    a_n = a
    b_n = b
    records = []

    for n in range(0, max_iter):
        m_n = (a_n + b_n)/2
        f_mn = f(m_n)
        records.append([n, a_n, b_n, m_n, f_mn])

        if abs(f_mn) < tol:
            break
        elif f(a_n)*f_mn < 0:
            b_n = m_n
        else:
            a_n = m_n

    table = pd.DataFrame(records, columns = ["n", "a_n", "b_n", "m_n", "f_mn"])
    return table

results = bisection_method(1, 1.5, 1e-5, 100)
print(results)


```
